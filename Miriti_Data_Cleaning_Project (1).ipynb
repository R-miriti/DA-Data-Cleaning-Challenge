```python
# Created a function to fetch unique characters

def get_unique_characters_in_column(df, column_name):
    # Initialize an empty set called unique_chars
    unique_chars = set()
    # iterate over each row in specified column
    for row in df[column_name]:
        # for each row, it converts each value to a string then adds it to the empty set above using the str()
        # converts that string to a set of characters using the set() 
        unique_chars.update(set(str(row)))
    return unique_chars
```

```python
# remove the tilde "~"
df["Contract"] = [i.replace ("~", "-") for i in df["Contract"]]
df["Contract"].head(2)
```

```python
# We'll create a function to define each player's contract

def process_contract(value):
    # contract has a start and end date
    if '-' in value:
        dates = value.split(' - ')
        contract_start = dates[0]
        contract_end = dates[1]
        return [contract_start, contract_end]
    # player has currently no contract
    elif value == 'Free':
        return ['free_player']
    # on loan contract
    else:
        loan_start = value[:-8]
        return ['on_loan', loan_start]    
```

```python
process_contract('Jan 4, 2021 On Loan')
```

```python
# Instantiate the above function
new_cont = df["Contract"].apply(process_contract)
new_cont.head()
```

After having applied the function, there will be three types of values:

a list with start and end years for contract
a list with only one element: 'free'
a list with first element equals to 'on_loan' and 2nd element: on_loan_from

```python
import numpy as np

df["on_loan_from"] = [i[1] if i[0] == 'on_loan' else np.nan for i in new_cont]
df["contract_started"] = [i[0] if len(i[0]) == 4 else np.nan for i in new_cont]
df["contract_ends"] = [i[1] if len(i[0]) == 4 else np.nan for i in new_cont]
df["has_contract"] = [bool(i) for i in df.contract_started]
```

```python
df["on_loan_from"].unique()
```

```python
# Let's now add our new contract columns into our clean df
# convert the dates to datetime 
# convert the 'has contract' to bool
df_clean_data["Contract_started"] = pd.Series(df.contract_started.astype ("datetime64"))
df_clean_data["Contract_ends"] = pd.Series(df.contract_ends.astype ("datetime64"))
df_clean_data["On_loan_from"] = pd.Series(df.on_loan_from.astype ("datetime64"))
df_clean_data["Has_contract"] = pd.Series(df.has_contract.astype (bool))
```

```python
# Let's come up with a function to convert the 'height' column into cm
def convert_height(x):
    # this looks at the rows with height in ft and inches
    if x[-1] == '"':
        x = x.replace('\"', "")
        foot = int(x[0]) *  30.48
        inch = int(x[2]) * 2.54
        return round(foot + inch)
    elif x[-1] == "m":
        return int(x[:-2])
        
```

```python
# Instantiate the function above
df.Height = df.Height.apply(convert_height)
```

```python
# Let's now add it into our clean df
df_clean_data["Height [cm]"] = pd.Series(df["Height"].astype("int64"))
df_clean_data.head(2)
```

Let's come up with a function to convert the lbs to kgs

```python
# Same function as height
def convert_weight (w):
    if 'lbs' in w:
        w = w.replace("lbs", "")
        to_kg = float(w) * 0.45359237
        w = int(to_kg)
        return w
    else:
        return int(w[:-2])
```

```python
# Lets clean the Value column

print(df["Value"].isnull().any())
print(df["Value"].isna().any())
```

```python
df["Value"].unique()
```

```python
# let's create a function to remove the "€" and "k" symbols

def convert_value(v):
    if "M" in v:
        v = v.replace("M", "")
        v = float(v[1:]) * 1_000_000
        return v
    elif "K" in v:
        v = v.replace("K", "")
        v = float(v[1:]) * 1000
        return v
    else:
        return v[1:]
```

```python
convert_value("€15K")
```

```python
# Instantiate the function

df["Value [€]"] = df["Value"].apply(convert_value)
df.head(2)
```
